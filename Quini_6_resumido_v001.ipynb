```python
def number_match(number, array, mode, id_lotery):
    
    #print(f'$array[$number][$mode]["matches"]// {number}-{mode}')
    #print(array[number][mode]['matches'])
    
    y_n=1
    array[number][mode]['matches']+=1
    try:
        array[number][mode]['freq']= round(array[number][mode]['matches']/id_lotery,2)
    except ZeroDivisionError:
        array[number][mode]['freq']= 0
    
    
    if array[number][mode]['streak'] <= 0:
        array[number][mode]['streak']=1
    else:
        array[number][mode]['streak']+=1
    
    
    
    last_match=id_lotery
    last_match_diff=0
    
    values_to_historic=(array[number][mode]['matches'], array[number][mode]['freq'], y_n, last_match, last_match_diff)
    array[number][mode]['historic'][id_lotery]=values_to_historic
```

```python
def not_match(serie, array,mode, id_lotery):  #completa el historial aunque no hay salido en la serie
    
    for item in array:
        number=array[item]['value']
        if mf.is_number_in(number, serie)==False:
            y_n=0
            try:
                array[number][mode]['freq']= round(array[number][mode]['matches']/id_lotery,4)
            except ZeroDivisionError:
                array[number][mode]['freq']= 0
            
           
            last_match_diff= id_lotery - (array[number][mode]['historic'][id_lotery-1][3])
            last_match=array[number][mode]['historic'][id_lotery-1][3]
            
            #last_match=array[number][mode]['historic'][id_lotery-1][3]
            values_to_historic=(array[number][mode]['matches'], array[number][mode]['freq'],y_n, last_match, last_match_diff)
            array[number][mode]['historic'][id_lotery]=values_to_historic
            
```

```python
def get_range_num_freq(range_rounds,serie_id, mode, dict, n_list):
    
    
    if range_rounds != 10 and range_rounds !=20 and range_rounds != 30 and range_rounds != 50 and range_rounds != 100:
        return print('Los rangos de frecuencia tienen que ser 10,20,30,50 o 100....!')
    
    id= serie_id
    prev= id-1 #la idea es que muestre la freq hasta ANTES del sorteo para el cual se juega
    series=[]
    matches={}
    #values=[]
    freq_index= "freq" + str(range_rounds)
    
    if id<= range_rounds:
        #print('Id menor al rango aplicab{e')
        for num in numbers:
            
            
            numbers[num][mode][freq_index][id]=('Na','Na','Na','Na','Na')
        
        return (f'{id}- ID No aplicable ==> $numbers[num][mode][freq_index][id]= {numbers[num][mode][freq_index][id]}')
    
    
    for num in n_list:
        matches[num]=0
    
    
    start= prev-range_rounds+1
    
    for item in range (start,id):
        serie=dict.loc[item, 'serie']
        series.append(serie)
        
    for s in series:
        #print(s)
        for num in matches:
            if mf.is_number_in(num, s) == True:
                    matches[num]+=1
            
    
    for number in numbers:
        #values.append(id) --> va al índica del diccionario...
        #values.append(matches[number])
        try:
            range_freq= matches[number] / len(series)
        except ZeroDivisionError:
            range_freq=0
        #values.append(range.freq)
        
  ####IMPORTANTE! => $prev es el nro de serie histórica, por regla de 3 se sacarían los matches esperados...      
        historic_freq= numbers[number][mode]['historic'][prev][1] #esto debe ser la frecuencia historica al sorteo anterior
        historic_matches=numbers[number][mode]['historic'][prev][0]
        diff_freq = round(range_freq - historic_freq,3)
        #values.append(diff_freq)
        expected_matches=int(round(((historic_matches*range_rounds)/prev),0))
        diff_expected=matches[number] - expected_matches
        
        numbers[number][mode][freq_index][id] = (matches[number], range_freq, diff_freq, expected_matches, diff_expected)
        #print(f' Nro: {number} --> {freq_index} => {numbers[number][mode][freq_index][id]}') -->OK
        
```

```python
def set_first_values(dict, id, modo,loteria, serie, suma):
    dict[id]={}
    dict[id]['ID']=id
    dict[id]['mode']=modo
    dict[id]['serie']=serie
    dict[id]['lotery']=loteria
    for i in range(len(serie)):
        index='n'+str(i+1)
        dict[id][index]=serie[i]
    #dict[id]['serie']=serie
    dict[id]['suma']=suma
```

```python
def append_for_df(dict,index):
    values=[]
    for key in dict[index]:
        
        values.append(dict[index][key])
        
        
    values.pop(2)
    values.pop(1)
    return values
```

```python
def get_group_elements(serie, group):
    count=0
    y_n=0
    for num in serie:
        if mf.is_number_in(num,group):
            count+=1
    
    if count>0:
        y_n=1
    
    return [count,y_n]
```

```python
def get_group_frequency(dataframe_rows, elements,serie_id):
    #dataframe_dict=> Total, Revancha, Segunda, SS
    
    #serie_id ==> debe ser el sorteo actual (a diferencia de la frecuencia con los nros o con las posiciones de ranking)
    #la explicación está en la iteración en el for...
    
    
    #matches refiere la celda de matches del grupo de la serie anterior
    count=0
    
    for i in range(1, serie_id):  #==> $i representa el dataframe_rows[1]
        serie= dataframe_rows[i]['serie']
        for num in serie:
            if mf.is_number_in(num, elements):
                count+=1
        
    freq=round((count/serie_id-1),3)
    return (count, freq)
```

```python
def repeated_numbers(dict,r):  #r es el intervalo de filas...
    results=[]
    fx_result=[]
    
    for i in range (len(dict)):
        if i==0 or (i-r) < 0:
            y_n=0
            count=0
            streak=-1
            results_to_append=[count, y_n ,streak]
            results.append(results_to_append)
            fx_result.append(results_to_append)
            results=[]
        else:
            count=get_group_elements(dict.loc[i, 'serie'], dict.loc[i-r, 'serie'])[0]
            y_n=get_group_elements(dict.loc[i, 'serie'], dict.loc[i-r, 'serie'])[1]
            
            if y_n==1:
                if streak > 0:
                    streak+=1
                elif streak < 0:
                    streak=1
            else:
                if streak > 0:
                    streak= (-1)
                elif streak < 0:
                    streak += (-1)
            
            results_to_append=[count, y_n ,streak]
            results.append(results_to_append)
            fx_result.append(results_to_append)
            results=[]
    
    #print(fx_result)
    return fx_result
```

```python
def new_repeated_numbers(row_dict, dataframe):
    
    
    
    row_id=row_dict['ID']
    serie=row_dict['serie']
    acumulate=0
    
    if row_id <= 1:
        return acumulate
    else:
        
        previous=dataframe.loc[row_id-2, 'serie']
        
        for x in serie:
            for y in previous:
                if x==y:
                    acumulate+=1
                    
        return acumulate
```

```python
def desv_in_serie(serie, dict):
    
    #streak=0
    #consecutives=0 redundante, len(consec_num)
    
    desv_indexes=['desv.1-2', 'desv.2-3', 'desv.3-4', 'desv.4-5','desv.5-6']
    
    #desv_indexes=['desv.0-1','desv.1-2', 'desv.2-3', 'desv.3-4', 'desv.4-5','desv.5-6', 'desv.6-45']
    normal=0
    big_desv=0
    big_desv_pos=[]
    consec_num=[]
    consec_groups=0
    atipical_desv=0
    
    for i in range (len(serie)):
        
        
        if i ==0:
            
            previous= serie[i]
            #index=desv_indexes[i]
        else:
            tuple_pair=(previous, serie[i])
            #index=desv_indexes[i-1]
            
            index=desv_indexes[i-1]

            div_big_huge=dict[index].quantile(0.75)+(dict[index].quantile(0.75)+(dict[index].quantile(0.75)/2))
            
            #print(row[index])
            if np.std((previous, serie[i])) == 0.5: #una desviación de 0.5 significa que los números son consecutivos
                
                if len(consec_num) ==0: #es el primer par de nros consecutivos
                    consec_num.append(previous)
                    consec_num.append(serie[i])
                    #consecutives=2
                    consec_groups+=1
                else:
                    if mf.is_number_in(previous, consec_num)==True: #si $previous ya está, significa que es un grupo de 3 o más
                        consec_num.append(serie[i])
                        #no se agrega otro par, es parte del mismo...
                    else: #es False... si $previous no está en el list, es un par de consecutivos nuevo
                        consec_num.append(previous)
                        consec_num.append(serie[i])
                        consec_groups+=1
                        
            elif np.std(tuple_pair) > dict[index].quantile(0.75) and np.std(tuple_pair) <= div_big_huge: #es un intervalo grando o uno enorme...         
                #if np.std(tuple_pair) <= div_big_huge:
                #print(f"la desv de la tupla ({tuple_pair}) es: {np.std(tuple_pair)} es mayor al quantil 75 ({dict[index].quantile(0.75)})")
                #print(f"... pero es menor o igual al valor de $div_big_huge: {div_big_huge}. Es un intervalo grande")
                big_desv+=1
                big_desv_pos.append(i)
                #big_desv_pos.append(index)
            
            elif np.std(tuple_pair) > dict[index].quantile(0.75) and np.std(tuple_pair) > div_big_huge:
                #if np.std((previous, serie[i])) >= 7.0:   ---> 4/2/23: No se pueden considerar valores absolutos ya que la magnitud
                                                            #de los intervalos cambia en cada caso
                #print(f"la desv de la tupla ({tuple_pair}) es: {np.std(tuple_pair)} es mayor al quantil 75 ({dict[index].quantile(0.75)})")
                #print(f"y es mayor al valor de $div_big_huge: {div_big_huge}. Es un intervalo enorme")
                atipical_desv +=1
            #elif np.std((previous, serie[i])) >= 7.0:
            #    print(f'Mayor o igual a 7.0: {np.std((previous, serie[i]))}')
            #    atipical_desv +=1
                
            #else:
                #print(f'NO es Mayor o igual a 7.0: {np.std((previous, serie[i]))}')
            
            elif np.std((previous, serie[i])) >= 1.0 and np.std((previous, serie[i])) <= dict[index].quantile(0.75): #el intervalo es normal...
                normal+=1

                
            previous= serie[i]

    results=[]
    results.append(len(consec_num)) #[0]--> cantidad de nros consecutivos
    results.append(consec_groups)  #[1] --> series de consecutivos... si es un par o es una serie de 4 consecutivos, o 2 pares...
    results.append(normal)          #[2]--> cantidad de intervalos normales en la serie (entre 1 y 9)* *9 es un nro dinámico...
    results.append(big_desv)        #[3]--> cantidad de intervalos fuera de lo normal...
    results.append(big_desv_pos)    #[4]--> es una lista... es para otra fx que determinen dónde están los intervalos
    results.append(atipical_desv)   #[5]--> señala cuantos intervalos atípicos hay
    return results
```
